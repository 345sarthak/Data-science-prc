```python
import pandas as pd
df=pd.DataFrame({'school_code':['s001','s002','s003','s001','s002','s004'],
      'Name':['Sarthak','Akash','Abhinay','Ritik','Kiran','Rohan'],
      'Class':['V','V','VI','VI','V','VI'],
      'Age':[20,21,23,23,34,76],
      'Height':[125,142,124,134,124,315],
      'Weight':[69,60,45,56,67,34],
      'Address':['Nagar','Panvel','Kamothe','Nerul','Pune','Takli']})
```

```python
df
```

```python
grouped_single=df.groupby("school_code").agg({'Age':['mean','min','max']})
grouped_single
```

```python
#Practical 5
multiple_groups=df.groupby(['school_code','Class']).agg({'Height':['mean','min','max'],'Weight':['sum','min','count']})
print(multiple_groups)
```
