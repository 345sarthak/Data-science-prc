```python
import seaborn as sns
sns.get_dataset_names()
```

```python
#Load and Display Head of Titanic Dataset
df=sns.load_dataset('titanic')
df.head()
```

```python
# Display Count of Null Values in Each Column of the DataFrame
df.isnull().sum()
```

```python
#To fill null values in the age column
mean_age=df['age'].mean()
df['Age Mean']=df['age'].fillna(mean_age)
df
```

```python
df.value_counts('deck').nunique()
```

```python
mode_deck=df['deck'].mode()[0]
df['deck mode']=df['deck'].fillna(mode_deck)
df
```

```python
mode_embarked=df['embarked'].mode()[0]
df['embarked mode']=df['embarked'].fillna(mode_embarked)
df
```

```python
mode_embark_town 	=df['embark_town'].mode()[0]
df['embark_town mode']=df['embark_town'].fillna(mode_embark_town 	)
df
```
